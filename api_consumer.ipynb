```python
!pip install stackapi
from stackapi import StackAPI
SITE = StackAPI('cseducators')
```

```python
posts = []

for i in range(74):
  delim = []
  for j in range(i*100,(i+1)*100):
    delim.append(f'{j}')
  posts.append(SITE.fetch(f"posts/{';'.join(delim)}"))

posts
```

```python
post_list = []

for query in posts:
  post_list.extend(query['items'])

post_list
```

```python
post_list[0].keys()
```

```python
import pandas as pd
```

```python
posts_dict = dict()


for post in post_list:
  id = post['post_id']
  post_clone = dict()

  for item in post.items():
    if item[0] == 'post_id':
      continue
    elif item[0] == 'owner':
      post_clone[item[0]] = (item[1]['user_id'], item[1]['display_name'])
    else:
      post_clone[item[0]] = item[1]
    
  posts_dict[id] = post_clone

posts_dict
```

```python
data = pd.DataFrame.from_dict(posts_dict, orient='index')
data.head()
```

```python
data.to_csv('post_csedu.csv')
```
